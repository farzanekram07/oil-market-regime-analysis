# 1. Business Context
### Why should anyone care about oil prices?

- Oil is one of the most strategically important commodities in the global economy. Changes in oil prices influence energy markets, industrial activity, transportation costs, and economic growth.

- For oil and gas companies, fluctuations in oil prices create uncertainty in operational and strategic planning. Decisions related to production, maintenance, inventory management, procurement, logistics, and budgeting are often influenced by expectations about future market conditions.

- As a result, __understanding the factors that drive oil price movements__ is essential for __supporting informed business decisions and managing market-related risks__.

# 2. Problem Statement
### What problem are we trying to solve? 

- Oil prices are influenced by a complex combination of supply, demand, economic conditions, geopolitical events, and policy decisions. These factors can create significant market volatility, making it difficult for organizations to plan operations effectively.

- Without a structured understanding of the drivers behind oil price movements, organizations may face challenges in production planning, maintenance scheduling, inventory management, procurement timing, and resource allocation.

- The __objective__ of this project is __to investigate the drivers__ of oil price movements and __develop an analytical framework__ capable of supporting business planning and decision-making

# 3. Connect market analysis to Business Value.
### Why This Problem Matters?

- Understanding oil price movements can support decision-making across multiple business functions.

- Production teams can use market insights to improve planning decisions.

- Maintenance teams can schedule operational activities during periods of lower business impact.

- Supply chain teams can better coordinate logistics and transportation requirements.

- Inventory managers can make more informed storage and stock management decisions.

- Financial teams can improve forecasting and budgeting processes.

- Overall, __by improving visibility into market conditions, organizations can move from reactive decision-making to proactive planning, reducing uncertainty and improving operational efficiency__.
